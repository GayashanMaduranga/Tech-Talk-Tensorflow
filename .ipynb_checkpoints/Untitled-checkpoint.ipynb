```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
sess = tf.InteractiveSession()
```

```python
writer = tf.summary.FileWriter('./output',sess.graph)
```

```python
x_data = np.linspace(0.0,10.0,1000)
```

Y = 0.5X + 5

```python
noise = np.random.randn(len(x_data))
```

```python
y_data = 0.5 * x_data + 5 + noise
```

```python
plt.plot(x_data,y_data,'o')
```

Y = Wx + b

```python
W = tf.Variable([1.23],dtype=tf.float32)
b = tf.Variable([-6],dtype=tf.float32)
```

```python
Y = tf.placeholder(tf.float32)
x = tf.placeholder(tf.float32)
```

```python
Linear_model = W*x + b 
```

```python
error = Y - Linear_model
cost = tf.square(error)/(2*len(x_data))
```

```python
loss = tf.reduce_sum(cost)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.01)
```

```python
train = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()
```

```python
sess.run(init)
```

```python
for i in range(10000):
    sess.run(train,{x:x_data,Y:y_data})
    print('Loss >>> ' , sess.run(loss,{x:x_data,Y:y_data}))
```

```python
c , d = sess.run([W,b])
```

```python
print(c,d)
```

```python
y_hat = c * x_data + d 
```

```python
plt.plot(x_data,y_data,'o')
plt.plot(x_data,y_hat,'r')
```

```python
writer.close()
```
