```python
x_data = np.linspace(0.0,10.0,100)
```

```python
noise = np.random.randn(len(x_data))
```

y = mx + b

```python
y_data = (0.5*x_data) + 5 + noise
```

```python
plt.plot(x_data,y_data,'o')
```

Y = Wx + b

```python
W = tf.Variable([1.92],dtype=tf.float32)
b = tf.Variable([2.121],dtype=tf.float32)
```

```python
Y = tf.placeholder(tf.float32)
x = tf.placeholder(tf.float32)
```

```python
linear_model = W*x + b
```

```python
error = Y - linear_model
## squared_error = tf.square(error)/(2*len(x_data))
squared_error = tf.square(error)
```

```python
loss = tf.reduce_mean(squared_error)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.01)
```

```python
train = optimizer.minimize(loss)
```

```python
sess = tf.InteractiveSession()
```

```python
init = tf.global_variables_initializer()
```

```python
sess.run(init)
```

```python
for i in range(10000):
    sess.run(train,{x:x_data,Y:y_data})
    print('Loss >>> ', sess.run(loss,{x:x_data,Y:y_data}))
```

```python
c , d = sess.run([W,b])
```

```python
y_hat = np.array(c * x_data + d)
```
